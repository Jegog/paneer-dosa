```python
import MeCab

f= open('pana_ar2023j_a4.txt', encoding="utf-8")
text = f.read()
f.close
```

```python
#stopwordsの指定
with open("Japanese.txt","r") as f:
    stopwords1 = f.read().split("\n")

stopwords2 =["以下","ため","当社","当行","場合","影響","可能性","経営",
            "業績","状況","グループ","こと","平成","年","月","期","当","他",
            "等","お","これ","a","b","c","%","以上","もの","株式会社",
            "経営方針","もの","とも","ある","よる","的","化","*"]
stopwords = list(set(stopwords1+stopwords2))
```

```python
import re

#特定の条件の除去
replaced_text = text.lower()
replaced_text = re.sub(r'[【】]', ' ', replaced_text)       # 【】の除去
replaced_text = re.sub(r'[（）()]', ' ', replaced_text)     # （）の除去
replaced_text = re.sub(r'[［］\[\]]', ' ', replaced_text)   # ［］の除去
replaced_text = re.sub(r'[@＠]\w+', '', replaced_text)  # メンションの除去
replaced_text = re.sub(r'\d+\.*\d*', '', replaced_text) #数字を0にする

# 辞書を指定してタガー生成
path = "-Ochasen -d /usr/lib/x86_64-linux-gnu/mecab/dic/mecab-ipadic-neologd"
#path = "-d /usr/lib/x86_64-linux-gnu/mecab/dic/mecab-ipadic-neologd"
tagger = MeCab.Tagger(path)

# 形態素解析結果を出力
for line in tagger.parse(replaced_text).splitlines():
    print(line)
```

```python
import re
from collections import Counter

# 辞書を指定してタガー生成
tagger = MeCab.Tagger("-Ochasen -d /usr/lib/x86_64-linux-gnu/mecab/dic/mecab-ipadic-neologd")
result = tagger.parse(text)
result_lines = result.split('\n')

result_words = []
words = []

for result_line in result_lines:
    result_words.append(re.split('[\t,]', result_line))
    
for result_word in result_words:

    if (    result_word[0] not in ('EOS', '')
        and (   ('名詞-一般') in result_word[3] 
             or ('名詞-サ変接続') in result_word[3] 
             or ('名詞-固有名詞') in result_word[3] 
            ) ):

            words.append(result_word[0])

# 頻出単語TOP10を抽出
counter = Counter(words)
for word, count in counter.most_common(300):
    print('%s : %s' % (word, count))
```
