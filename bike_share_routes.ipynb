```python
import pandas as pd
import matplotlib.pyplot as plt

# データを読み込む
data = pd.read_csv("trip_od_202006_202105.csv")  # データファイルのパスを指定する

# 直線の始点と終点の座標を計算する
data['start_point'] = list(zip(data['start_lat'], data['start_lng']))
data['end_point'] = list(zip(data['end_lat'], data['end_lng']))

# データをride_countでソートする
data.sort_values(by='ride_count', ascending=False, inplace=True)

# グラフを描画する
plt.figure(figsize=(10, 8))
for i, row in data.iterrows():
    plt.plot([row['start_lat'], row['end_lat']], [row['start_lng'], row['end_lng']], 
             linewidth=row['ride_count']/1000, color='blue')

plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Bike Share Routes')
plt.grid(True)
plt.show()
```
