```python
#テキストファイルを一行ごとに読み込んで、記載された名称のフォルダを作成
import os

# テキストファイルのパス
file_path = 'data/folder_names.txt'

# フォルダを作成したい場所のパス
output_dir = 'ir/2024'  # 必要に応じて指定する

# 出力フォルダが存在しない場合は作成
os.makedirs(output_dir, exist_ok=True)

# テキストファイルを開いて各行を読み込む
with open(file_path, 'r', encoding='utf-8') as f:
    for line in f:
        folder_name = line.strip()  # 行末の改行などを除去
        if folder_name:  # フォルダ名が空でない場合のみ処理
            # フォルダパスを作成
            folder_path = os.path.join(output_dir, folder_name)
            # フォルダを作成（存在しない場合のみ）
            os.makedirs(folder_path, exist_ok=True)
            print(f"フォルダ '{folder_name}' を作成しました。")
```
