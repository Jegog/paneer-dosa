```python
import csv

# CSVファイルのパス
csv_file_path = 'page_info.csv'


# 読み込んだデータを格納するリスト
data = []

# CSVファイルを読み込む
with open(csv_file_path, newline='', encoding='utf-8-sig') as csvfile:
    csvreader = csv.reader(csvfile)
    #一行目にタイトルを入れるときは一行目をスキップ
    #next(csvreader)
    for row in csvreader:
        # ページに関する情報のフィールドをカンマで分割してリストにする
        #Sustainability含む
        A_list = [int(item) for item in row[3].strip('"').split(',')]
        #Sustainability除く
        B_list = [int(item) for item in row[4].strip('"').split(',')]
        #IP関連
        IP_list = [int(item) for item in row[5].strip('"').split(',')if item != '-']
        
        # サブリストを含めて新しいリストを作成
        processed_row = row[:3] + [A_list] + [B_list] + [IP_list] + row[6:]
        # 格納用リストに追加
        data.append(processed_row)

# 結果を表示
for row in data:
    print(row[6])
    print('----')
```

```python
import pandas as pd

# CSVファイルのパス
csv_file_path = 'data/page_info.csv'

# CSVファイルを読み込む
df_pageinfo = pd.read_csv(csv_file_path, encoding='utf-8-sig')

# フィールドをリストに分割して数値に変換し、アンダースコアを除外する関数
def convert_to_list(cell):
    return [int(x) for x in cell.strip('"').split(',') if x != '-' and x.strip() != '']

# 範囲1をリストに変換
df_pageinfo['ページ範囲1'] = df_pageinfo.iloc[:, 3].apply(convert_to_list)

# 範囲2をリストに変換
df_pageinfo['ページ範囲2'] = df_pageinfo.iloc[:, 4].apply(convert_to_list)

# 範囲3をリストに変換
df_pageinfo['page_area_3'] = df_pageinfo.iloc[:, 5].apply(convert_to_list)

df_pageinfo
```
