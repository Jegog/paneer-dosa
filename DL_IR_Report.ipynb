```python
import os
import pandas as pd
import requests

# CSVファイルのパス
csv_file_path = 'data/URL_list.csv'

# CSVを読み込む
df = pd.read_csv(csv_file_path)

# ダウンロードエラーを記録するリスト
download_errors = []

# 各行について処理
for _, row in df.iterrows():
    # フォルダ名は'ir/'+Code
    folder_name = 'ir/'+str(row['Code'])+ '_' + str(row['CompanyName'])
    
    # フォルダがなければ作成
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)
    
    # URL1
    if pd.notna(row['URL1']):
        try:
            # ファイル名はURLから抽出する
            file_name = os.path.basename(row['URL1'])
            file_path = os.path.join(folder_name, file_name)
            
            # PDFをダウンロードして保存
            response = requests.get(row['URL1'])
            if response.status_code == 200:
                with open(file_path, 'wb') as f:
                    f.write(response.content)
            else:
                download_errors.append(row['URL1'])
        except Exception as e:
            download_errors.append(row['URL1'])
    
    # URL2
    if pd.notna(row['URL2']):
        try:
            file_name = os.path.basename(row['URL2'])
            file_path = os.path.join(folder_name, file_name)
            
            response = requests.get(row['URL2'])
            if response.status_code == 200:
                with open(file_path, 'wb') as f:
                    f.write(response.content)
            else:
                download_errors.append(row['URL2'])
        except Exception as e:
            download_errors.append(row['URL2'])

# ダウンロードエラーがある場合、出力
if download_errors:
    print("ダウンロードエラーのあったURL:")
    for error in download_errors:
        print(error)
```
