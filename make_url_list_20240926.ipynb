```python
# HP上で公開されている統合報告書のリンク一覧を取得・リスト化

import requests
from bs4 import BeautifulSoup
import pandas as pd

# 企業コード
company_code = 25030
# 企業名
company_name = 'キリンホールディングス'
# 発行年月
release = 202406
# 指定したURL
url = 'https://www.kirinholdings.com/jp/investors/library/integrated/'

# URLにアクセスしてHTMLを取得
response = requests.get(url)

# 文字コードをUTF-8に指定
response.encoding = 'utf-8'

# BeautifulSoupでHTMLを解析
soup = BeautifulSoup(response.text, 'html.parser')

# 特定のdiv要素を取得（classとdata-fixed属性で指定）
#target_div = soup.find('div', {'class': 'content-fixed', 'data-fixed': ''})
target_div = soup.find('div', {'class': 'content-fixed'})

# リンクを格納するリスト
link_data = []

# 取得したdiv内のaタグをすべて抽出
if target_div:
    links = target_div.find_all('a')  # div内のaタグを全て取得
    for link in links:
        href = link.get('href')  # リンクのURLを取得
        text = link.get_text(strip=True)  # リンクのテキストを取得
        if href:
            link_data.append([text, href])  # テキストとリンクをリストに格納
else:
    print('指定したdivが見つかりませんでした。')

# リンクのテキストとURLをデータフレームに変換
df = pd.DataFrame(link_data, columns=['text', 'url'])

df['code'] = company_code
df['company'] = company_name
df['release'] = release
df['chk'] = ''#リストからデータを取得するかのチェック項目
df=df[['code','company','release','text','url','chk']]
# データフレームを表示
print(df)

file_name_url = f"output/url_list_{company_code}_{company_name}_{release}.csv"
df.to_csv(file_name_url, encoding="utf_8_sig", index=False)    
```
